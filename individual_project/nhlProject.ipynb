```python
import re
from utils.soup import getSoup
import matplotlib as plt
from matplotlib import pyplot
import pandas as pd
import numpy as np
import seaborn as sns

page = 'https://www.espn.com/nhl/stats/team/_/view/goaltending'
soup  = getSoup(page)
```

```python
names_original = soup.find_all(class_ = re.compile('flex items-start mr7'))
team_names = []
for i in names_original:
    temp = i.text.strip()
    team_names.append(temp)

print(team_names)
```

```python
a = soup.find_all(class_='Table__ScrollerWrapper relative overflow-hidden')
stats = dict()
stats['Categories'] = ['GP', 'GA/G','W','L','OTL','SA','GA','S','SV%','SO','SOSA','SOS','SOS%']
for name in team_names:
    stats[name] = 0
print(stats)
count = 0
for i in a:
    b = i.find_all(class_ = 'Table__Scroller')
    for k in b:
        c = k.find_all(class_ = 'Table__TBODY')
        for l in c:
            d = l.find_all(class_ = 'Table__TR Table__TR--sm Table__even')
            
            for p in d:
                temp = []
                e = p.find_all(class_ = 'Table__TD')
                for stat in e:
                    s = stat.text.strip()
                    new_string = re.sub(",", "", s)
                    temp.append(float(new_string))
                stats[team_names[count]] = temp
                count = count + 1
print(stats)
```

```python
for i, j in stats.items():
  print(i, ' : ', j)
```

```python
page2 = 'https://www.espn.com/nhl/stats/team'
soup2  = getSoup(page2)
names_original = soup2.find_all(class_ = re.compile('flex items-start mr7'))
team_names2 = []
for i in names_original:
    temp = i.text.strip()
    team_names2.append(temp)

print(team_names2)
```

```python
a = soup2.find_all(class_='Table__ScrollerWrapper relative overflow-hidden')
stats['Categories'] = ['GP', 'GA/G','W','L','OTL','SA','GA','SV','SV%','SO','SOSA','SOS','SOS%','GP2','GF/G','A','PTS','PPG','PP%','SHG','S','S%','PIM','PK%','SOA','SOG','SO%']
count = 0
for i in a:
    b = i.find_all(class_ = 'Table__Scroller')
    for k in b:
        c = k.find_all(class_ = 'Table__TBODY')
        for l in c:
            d = l.find_all(class_ = 'Table__TR Table__TR--sm Table__even')
            for p in d:
                temp = []
                e = p.find_all(class_ = 'Table__TD')
                for stat in e:
                    s = stat.text.strip()
                    new_string = re.sub(",", "", s)
                    temp.append(float(new_string))
                stats[team_names2[count]] += temp
                count = count + 1
print(stats)
```

```python
df = pd.DataFrame(data = stats)
df = df.set_index('Categories')
df = df.transpose()
df['W%'] = df['W'] / (df['W'] + df['L'])
df = df.drop(columns = ['GP2','W', 'L'])
df
```
